# SETUP

```python
import numpy as np
from probe_plotting2 import plot_one_line, plot_one_scatter
import matplotlib
```

```python
matplotlib.rc('text', usetex=True)
matplotlib.rcParams.update({'figure.autolayout': True, 'text.latex.unicode': True})
```

```python
%matplotlib inline
```

# GLOBAL OPTIONS

```python
line_width=2
pad=12
ticks_font_size=40
label_font_size=40
legend_font_size=40
border_thickness=2.0
ticks_length=5.0
ticks_width=2.0
size=25
```

# SAMPLE DATA

```python
x = np.linspace(0.0, 1.0, 100)
y1 = x**2
y2 = x**3
y3 = x**4
```

# PLOT_ONE_LINE

```python
fig = plot_one_line(x, y1, line_color='green', line_width=line_width,
                    data_label=r'$y_\mathrm{1}$')
fig = plot_one_line(x, y2, line_color='red', line_width=line_width,
                    data_label=r'$y_\mathrm{2}$', y_secondary=True, fig=fig)

fig = plot_one_line(x, y3, line_color='blue', line_width=2, fig=fig,
                    x_label=r'$x$', y_label =r'$y_{\mathrm{1,3}}$',
                    pad=pad, ticks_font_size=ticks_font_size,
                    label_font_size=label_font_size, legend_font_size=legend_font_size,
                    border_thickness=border_thickness, ticks_length=ticks_length,
                    ticks_width=ticks_width, show=True, y2labelpad=15,
                    data_label=r'$y_\mathrm{3}$', y2_label=r'$y_\mathrm{2}$')

fig['fig'].savefig('plot_three_lines.pdf')
```

# PLOT_ONE_SCATTER

```python
fig = plot_one_scatter(x, y1, line_color='green', fill_color='green', data_label=r'$y_\mathrm{1}$',
                       size=size, alpha=0.6)
fig = plot_one_scatter(x, y2, line_color='red', fill_color='red', data_label=r'$y_\mathrm{2}$',
                       size=12, fig=fig, err_y=np.ones(y2.size)*0.1*np.linspace(0.0, 1.0, y2.size), alpha=0.6,
                       err_line_width=2.0, err_capthick=2.0, err_capsize=2.0)
fig = plot_one_scatter(x, y3, line_color='blue', fill_color='blue', fig=fig,
                       x_label=r'$x$', y_label =r'$y_{\mathrm{1,3}}$',
                       pad=pad, ticks_font_size=ticks_font_size,
                       label_font_size=label_font_size, legend_font_size=legend_font_size,
                       border_thickness=border_thickness, ticks_length=ticks_length,
                       ticks_width=ticks_width, show=True, data_label=r'$y_\mathrm{3}$',
                       size=size, legend_loc=2, xlim=(0.0,1.0), alpha=0.6)

fig['fig'].savefig('plot_three_scatters.pdf')
```
